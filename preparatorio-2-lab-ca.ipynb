```python
import cmath
import numpy
import math
from cmath import phase 
import sympy as sym
from numpy import exp, abs, angle
import numpy as np

#variaveis simbolicas para as tensões
Vb, Vc, Vd = sym.symbols('Vb, Vc, Vd')

#reatâncias
Xc = -1/(2*math.pi*60*500*10**(-6))
Xl1 = 2*math.pi*60*100*10**(-6)
Xl2 = 2*math.pi*60*200*10**(-6)

#impedancias do circuito
Z1 = complex(10,Xl1)
Z2 = complex(0,Xc)
Z3 = complex(2,Xl2)
Z4 = complex(0,Xc)
Z5 = complex(1,Xl1)
Z6 = complex(0,Xc)
Z7 = complex(1,0)

impedancias = [Z1, Z2, Z3, Z4, Z5, Z6, Z7]

#tensao da fonte
V1 = complex(179.6,0)

#calculo da Zeq
Zp1 = Z1*Z2/(Z1 + Z2)
Zp2 = Zp1 + Z3
Zp3 = Zp2*Z4/(Zp2 + Z4)
Zp4 = Zp3 + Z5
Zp5 = Zp4*Z6/(Zp4 + Z6)
Zp6 = Zp5 + Z7

#corrente total
It = V1/Zp6
print("\n modulo da corrente total é: {:.2f} e seu angulo é: {:.2f}" .format(cmath.polar(It)[0], math.degrees(cmath.phase(It))))

#fator de potencia
fp = math.cos(phase(Zp6))
print("\n fator de potência do circuito: {:.2f} capacitivo \n".format(fp))
print("\n modulo da impedância complexa é: {:.2f} ohm e seu angulo é: {:.2f} \n".format(cmath.polar(Zp6)[0],math.degrees(cmath.polar(Zp6)[1])))

#potencia aparente total
St = (V1/np.sqrt(2))*(np.conj(It)/np.sqrt(2))
print("\n potencia ativa total: {:.2f}W potencia reativa total: {:.2f}VAr" .format(St.real,St.imag))

#equacoes nodais
equacoes = []
#no b
equacoes.append(-((V1 - Vb)/Z7) + (Vb - Vc)/Z5 + Vb/Z6)
#no c
equacoes.append(-((Vb - Vc)/Z5) + Vc/Z4 + (Vc-Vd)/Z3)
#no d
equacoes.append(-((Vc-Vd)/Z3) + Vd/Z2 + Vd/Z1)

incognitas =[Vb, Vc, Vd]

#resolvendo equacoes nodais
solucao = sym.solve(equacoes,incognitas)

#correntes atravessando as impedâncias
I1 = (solucao[Vb] - solucao[Vc])/Z5
I2 = solucao[Vb]/Z6
I3 = solucao[Vc]/Z4
I4 = (solucao[Vc]-solucao[Vd])/Z3
I5 = solucao[Vd]/Z2
I6 = solucao[Vd]/Z1

print("\n módulo de tensão nó b: {:.2f} V".format(sym.Abs(solucao[Vb])))
print("\n ângulo da tensão do nó b: {:.2f}\n".format(math.degrees(sym.arg(solucao[Vb]))))
print("\n módulo de tensão nó c: {:.2f} V".format(sym.Abs(solucao[Vc])))
print("\n ângulo da tensão do nó c: {:.2f}\n".format(math.degrees(sym.arg(solucao[Vc]))))
print("\n módulo de tensão nó d: {:.2f} V".format(sym.Abs(solucao[Vd])))
print("\n ângulo da tensão do nó d: {:.2f}\n".format(math.degrees(sym.arg(solucao[Vd]))))

print("\n módulo da corrente na impedância 6: {:.2f} A".format(sym.Abs(I1)))
print("\n ângulo da corrente na impedância 6: {:.2f}\n".format(math.degrees(sym.arg(I1))))
print("\n módulo da corrente na impedância 5: {:.2f} A".format(sym.Abs(I2)))
print("\nângulo da corrente na impedância 5: {:.2f}\n".format(math.degrees(sym.arg(I2))))
print("\n módulo da corrente na impedância 4: {:.2f} A".format(sym.Abs(I3)))
print("\nângulo da corrente na impedância 4: {:.2f}\n".format(math.degrees(sym.arg(I3))))
print("\n módulo da corrente na impedância 3: {:.2f} A".format(sym.Abs(I4)))
print("\nângulo da corrente na impedância 3: {:.2f}\n".format(math.degrees(sym.arg(I4))))
print("\n módulo da corrente na impedância 2: {:.2f} A".format(sym.Abs(I5)))
print("\nângulo da corrente na impedância 2: {:.2f}\n".format(math.degrees(sym.arg(I5))))
print("\n módulo da corrente na impedância 1: {:.2f} A".format(sym.Abs(I6)))
print("\nângulo da corrente na impedância 1: {:.2f}\n".format(math.degrees(sym.arg(I6))))

#potencias aparentes das impedâncias
S1 = (solucao[Vd]/np.sqrt(2))*(np.conj(I6)/np.sqrt(2))
S2 = (solucao[Vd]/np.sqrt(2))*(np.conj(I5)/np.sqrt(2))
S3 = ((solucao[Vc]-solucao[Vd])/np.sqrt(2))*(np.conj(I4)/np.sqrt(2))
S4 = (solucao[Vc]/np.sqrt(2))*(np.conj(I3)/np.sqrt(2))
S5 = ((solucao[Vb]-solucao[Vc])/np.sqrt(2))*(np.conj(I1)/np.sqrt(2))
S6 = (solucao[Vb]/np.sqrt(2))*(np.conj(I2)/np.sqrt(2))
S7 = ((V1-solucao[Vb])/np.sqrt(2))*(np.conj(It)/np.sqrt(2))

print("\n potência ativa na impedância 7: {:.2f} W\n \n potência reativa na impedância 7: {:.2f} VAr\n".format(sym.re(S7), sym.im(S7)))
print("\n potência ativa na impedância 6: {:.1} W\n \n potência reativa na impedância 6: {:.2f} VAr\n".format(str(sym.re(S6)), sym.im(S6)))
print("\n potência ativa na impedância 5: {:.2f} W\n \n potência reativa na impedância 5: {:.2f} VAr\n".format(sym.re(S5), sym.im(S5)))
print("\n potência ativa na impedância 4: {:.2} W\n \n potência reativa na impedância 4: {:.2f} VAr\n".format(str(sym.re(S4)), sym.im(S4)))
print("\n potência ativa na impedância 3: {:.2f} W\n \n potência reativa na impedância 3: {:.2f} VAr\n".format(sym.re(S3), sym.im(S3)))
print("\n potência ativa na impedância 2: {:.2} W\n \n potência reativa na impedância 2: {:.2f} VAr\n".format(str(sym.re(S2)), sym.im(S2)))
print("\n potência ativa na impedância 1: {:.2f} W\n \n potência reativa na impedância 1: {:.2f} VAr\n".format(sym.re(S1), sym.im(S1)))
```
