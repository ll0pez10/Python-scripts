```python
import cmath
import numpy
import math
from math import pi
from cmath import phase 
import sympy as sym
from numpy import exp, abs, angle
import numpy as np

def paraleloImpedancia(list, num_imp):
    Zp = 0
    for i in range(num_imp):
        Zp = Zp + (1/list[i])
    return (1/Zp)

impedancias = []

Xc = -1/(2*pi*60*25*10**(-6))

Z=complex(100,Xc)

impedancias.append(Z)
impedancias.append(Z)
impedancias.append(Z)

print("*-------- circuito equilibrado em Y --------*")

#tensoes de linha
Vl1 = complex(220,0)
Vl2 = complex(-110,190.5256)
Vl3 = complex(-110,-190.5256)

#modulo do numero complexo
abs(Vl1)
#angulo numero complexo
math.degrees(phase(Vl2))

#tensoes de fase
Vf1 = cmath.rect(220/np.sqrt(3),(math.pi)/6)
Vf2 = cmath.rect(220/np.sqrt(3), ((2*math.pi)/3)+((math.pi)/6))
Vf3 = cmath.rect(220/np.sqrt(3), ((-2*math.pi)/3)+((math.pi)/6))

#correntes nas fases
IA = Vf1/Z
IB = Vf2/Z
IC = Vf3/Z
print(" corrente A:{:.2f} \n \n corrente B:{:.2f} \n \n corrente C:{:.2f} \n" . format(IA, IB, IC))

#potencia em cada ramo
S1 = Vf1*np.conj(IA)
S2 = Vf2*np.conj(IB)
S3 = Vf3*np.conj(IC)
#potencia ativa e reativa total
St = S1 + S2 + S3
print("A potencia total é : {:.2f} \n" .format(St))
#fator de potencia
fp = St.real/abs(St)
print("O fator de potencia desse circuito é: {:.2f}" . format(fp))
```

```python
print("*-------- circuito equilibrado Y triângulo --------*")

impedancias = []

Xc = -1/(2*math.pi*60*25*10**(-6))

Z=complex(100,Xc)

#transformação de impedâncias delta para estrela
impedancias.append(Z/3)
impedancias.append(Z/3)
impedancias.append(Z/3)

#tensoes de linha
Vl1 = complex(220,0)
Vl2 = complex(-110,190.5256)
Vl3 = complex(-110,-190.5256)

#tensoes de fase
Vf1 = cmath.rect(220/np.sqrt(3),(math.pi)/6)
Vf2 = cmath.rect(220/np.sqrt(3), ((2*math.pi)/3)+((math.pi)/6))
Vf3 = cmath.rect(220/np.sqrt(3), ((-2*math.pi)/3)+((math.pi)/6))

#correntes nas fases
IA = Vf1/impedancias[0]
IB = Vf2/impedancias[0]
IC = Vf3/impedancias[0]
print(" corrente A:{:.2f} \n \n corrente B:{:.2f} \n \n corrente C:{:.2f} \n" . format(IA, IB, IC))

#correntes de linha
IlA = cmath.rect(abs(IA)*np.sqrt(3), phase(IA) + (math.pi/6))
IlB = cmath.rect(abs(IB)*np.sqrt(3), phase(IB) + (math.pi/6))
IlC = cmath.rect(abs(IC)*np.sqrt(3), phase(IC) + (math.pi/6))
print(" corrente linha A:{:.2f} \n \n corrente linha B:{:.2f} \n \n corrente linha C:{:.2f} \n" . format(IlA, IlB, IlC))

#potencia em cada ramo
S1 = Vf1*np.conj(IA)
S2 = Vf2*np.conj(IB)
S3 = Vf3*np.conj(IC)
#potencia ativa e reativa total
St = S1 + S2 + S3
print("A potencia total é : {:.2f} \n" .format(St))
#fator de potencia
fp = St.real/abs(St)
print("O fator de potencia desse circuito é: {:.2f}" . format(fp))
```
