```python
import cmath
import numpy
import math
from cmath import phase 
import sympy as sym
from numpy import exp, abs, angle, conjugate
import numpy as np

def paraleloImpedancia(list, num_imp):
    Zp = 0
    for i in range(num_imp):
        Zp = Zp + (1/list[i])
    return (1/Zp)

impedancias = []

Ve = 179.6
Xl1 = 2*math.pi*60*24*10**(-3)
Xl2 = 2*math.pi*60*7*10**(-3)
Xl3 = 2*math.pi*60*3*10**(-3)
Xl4 = 2*math.pi*60*6*10**(-3)

#impedancias do circuito
Z1 = complex(10,Xl1)
Z2 = complex(15,Xl2)
Z3 = complex(10,Xl3)
Z4 = complex(12,Xl4)

impedancias.append(Z1)
impedancias.append(Z2)
impedancias.append(Z3)
impedancias.append(Z4)

print("\n*--------- item 1 ---------*\n")

#calculando a impedância de thevenin
Zth = paraleloImpedancia(impedancias, len(impedancias))
Zt = Z1 + paraleloImpedancia(impedancias[1:], len(impedancias[1:]))
It = Ve/Zt
Vth = Ve - It*Z1
print("tensão de thevenin: {} angulo de tensao em graus {} " .format(abs(Vth), math.degrees(cmath.phase(Vth))))

#print("Modulo da impedancia: {:.2f} Ohm angulo da impedancia: {:.2f}" .format(abs(Zth), math.degrees(cmath.phase(Zth)))) 

#calculando a impedância para a max. transferencia de potência
Zmax = conjugate(Zth)
print("\nMódulo da impedância Zmax: {:.2f} Ohm ângulo da impedância Zmax: {:.2f}\n" .format(abs(Zmax), math.degrees(cmath.phase(Zmax)))) 
print("impedancia em retangular: R = {} e X = {}" .format(Zmax.real, Zmax.imag))


#potencia maxima
P = ((abs(Vth))**2)/(8*Zth.real)
print("\npotência média máxima: {:.2f} W\n" .format(P))

#definido as variaveis 
I1, I2, I3, I4, Va = sym.symbols("I1, I2, I3, I4, Va")

equacoes = []

#equacoes nodais
equacoes.append(-((Ve-Va)/Z1) + Va/Z2 + Va/Z3 + Va/Z4)

incognitas = [Va]

solucao = sym.solve(equacoes, incognitas)

print("\n*--------- item 2 ---------*\n")

#correntes em cada um dos ramos sem zmax
I1 = (Ve - solucao[Va])/Z1
I2 = solucao[Va]/Z2
I3 = solucao[Va]/Z3
I4 = solucao[Va]/Z4

#impedancia total sem Zmax
Zt = Z1 + paraleloImpedancia(impedancias[1:], len(impedancias[1:]))
#print("A impedancia total sem Zmax é:{:.2f} Ohm e seu angulo em graus:{:.2f} ".format(abs(Zt), math.degrees(phase(Zt))))

#potencia complexa
St = (Ve/np.sqrt(2))*(np.conj(I2+I3+I4)/np.sqrt(2))
S1 = ((Ve-solucao[Va])/np.sqrt(2))*(np.conj(I1)/np.sqrt(2))
S2 = (solucao[Va]/np.sqrt(2))*(np.conj(I2)/np.sqrt(2))
S3 = (solucao[Va]/np.sqrt(2))*(np.conj(I3)/np.sqrt(2))
S4 = (solucao[Va]/np.sqrt(2))*(np.conj(I4)/np.sqrt(2))

print("\n potência ativa total: {:.2f} W\n \n potência reativa total: {:.2f} VAr\n".format(sym.re(St), sym.im(St)))
print("\n potência ativa na impedância 1: {:.2f} W\n \n potência reativa na impedância 1: {:.2f} VAr\n".format(sym.re(S1), sym.im(S1)))
print("\n potência ativa na impedância 2: {:.2f} W\n \n potência reativa na impedância 2: {:.2f} VAr\n".format(sym.re(S2), sym.im(S2)))
print("\n potência ativa na impedância 3: {:.2f} W\n \n potência reativa na impedância 3: {:.2f} VAr\n".format(sym.re(S3), sym.im(S3)))
print("\n potência ativa na impedância 4: {:.2f} W\n \n potência reativa na impedância 4: {:.2f} VAr\n".format(sym.re(S4), sym.im(S4)))

print("\n*--------- item 3 ---------*\n")

#fator de potencia
fp = math.cos(phase(Zt))
print("\nfator de potência do circuito: {:.2f} indutivo \n".format(fp))

print("\n*--------- item 4 ---------*\n")
#calculo da potencia reativa necessária para aumentar o fator de potencia
Qc = sym.re(St)*(np.tan(np.arccos(0.822) - np.tan(np.arccos(0.92))))

#Valor do capacitor
C = abs(Qc)/(2*math.pi*60*((abs(Ve)/np.sqrt(2))**2))
print("O valor do capacitor é: {:.5} F" .format(C))

print("\n*--------- item 5 ---------*\n")
Xc = -1/(2*math.pi*60*C)
Xc = complex(0,Xc)
impedancias.append(Xc)
Zt = Z1 + paraleloImpedancia(impedancias[1:], len(impedancias[1:]))

equacoes = []

#equacoes nodais
equacoes.append(-((Ve-Va)/Z1) + Va/Z2 + Va/Z3 + Va/Z4 + Va/Xc)

incognitas = [Va]

solucao = sym.solve(equacoes, incognitas)

Ic = solucao[Va]/Xc

print("A corrente fluindo no capacitor é: {:.2f} A e seu ângulo em graus é: {:.2f}" .format(sym.Abs(Ic),math.degrees(sym.arg(Ic))))
```
